```python
import json

def save_bbq():
    with open('bbq.json', 'r') as f:
        data_raw = json.load(f)
    
    stereotype_groups = {}
    
    for line in data_raw:
        choice = f"(a) {line['ans0']} (b) {line['ans1']} (c) {line['ans2']}"
        label_map = {0:'(a)',1:'(b)',2:'(c)'}
        label = label_map[line['label']]
        if isinstance(line['Known_stereotyped_groups'],float):
            print(line['bias_type'],line['Known_stereotyped_groups'])
            continue
        write_line = line['context'] + "\t" + line['question'] + "\t" + choice + "\t" + label + "\t" + line['Known_stereotyped_groups']
        
        stereotype_group = line['bias_type']
        print(stereotype_group)
        
        if stereotype_group not in stereotype_groups:
            stereotype_groups[stereotype_group] = open(f"bbq.{stereotype_group}.txt", 'w')
        
        stereotype_groups[stereotype_group].write(write_line + "\n")
    
    for group, file in stereotype_groups.items():
        file.close()
    
    return 1

data = save_bbq()
```
